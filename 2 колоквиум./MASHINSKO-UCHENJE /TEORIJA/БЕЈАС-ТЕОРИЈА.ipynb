### Зошто наивниот Беасов алгоритам се нарекува "наивен", а зошто "Баесов"?

```python
import requests
from bs4 import BeautifulSoup

# URL of the Medium article
url = "https://medium.com/@kirudang/why-naive-bayes-is-called-naive-and-what-are-the-benefits-of-being-naive-180757155b69"

# Fetch the article
response = requests.get(url)
if response.status_code == 200:
    # Parse the content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extract the main content
    # Medium's articles are often inside <article> or other specific tags
    article_content = soup.find('article')
    if article_content:
        print(article_content.text[:1000])  # Print the first 1000 characters for preview
    else:
        print("Couldn't find the article content!")
else:
    print(f"Failed to fetch the article. Status code: {response.status_code}")
```
